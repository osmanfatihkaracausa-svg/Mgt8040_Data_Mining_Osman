```python
# Cost Habits Tutorial
# Step 1: Import Requried Libraries and Load Data The two CSV files


import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#load the data

habits = pd.read_csv("../Data/Personal_Habits.csv")
spending = pd.read_csv("../Data/Daily_Spending.csv")



#Check the shape of the data
print("Personal Habits shape:", habits.shape)
print("Spending Habits shape:", spending.shape)

#Preview the Data
display(habits.head())
display(spending.head())
```

```python
# Step 2: Check column names and data types

print("Personal Habits columns:")
print(habits.columns)

print("\nDaily Spending columns:")
print(spending.columns)

print("\nPersonal Habits data types:")
print(habits.dtypes)

print("\nDaily Spending data types:")
print(spending.dtypes)
```

```python
# Convert Date columns to datetime format

habits["Date"] = pd.to_datetime(habits["Date"])
spending["Date"] = pd.to_datetime(spending["Date"])

print(habits.dtypes)
print(spending.dtypes)
```

```python
# Perform Left Join on Date

merged = pd.merge(
    habits,
    spending,
    on="Date",
    how="left"
)

print("Merged Shape:", merged.shape)

display(merged.head(15))
```

```python
# Replace missing values with 0

merged["Amount"] = merged["Amount"].fillna(0)
merged["Joy_Rating"] = merged["Joy_Rating"].fillna(0)

# Check remaining missing values

print(merged[["Amount", "Joy_Rating"]].isnull().sum())
```

```python
# Generate Correlation Matrix

corr = merged.corr(numeric_only=True)

corr
```

```python
# Create Sleep Group

merged["Sleep_Group"] = merged["Sleep_Hours"].apply(
    lambda x: "Above_7" if x >= 7 else "Below_7"
)

# Compare coffee and focus by sleep group

interaction = merged.groupby("Sleep_Group")[["Coffee_Cups", "Work_Focus"]].mean()

interaction
```
