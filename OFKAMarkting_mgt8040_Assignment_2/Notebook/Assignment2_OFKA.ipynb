```python
import pandas as pd

files = [
    "wc_traffic.csv",
    "ig_views.csv",
    "ig_interactions.csv",
    "ig_link clicks.csv",
    "ig_profile_visits.csv",
    "ig_follows.csv",
    "fb_views.csv",
    "fb_interactions.csv",
    "fb_link clicks.csv",
    "fb_visits.csv",
    "fb_follows.csv"
]

for file in files:
    print("\n" + "="*80)
    print(f"FILE: {file}")
    print("="*80)

    try:
        if file == "wc_traffic.csv":
            df = pd.read_csv(
                f"../Data/{file}",
                skiprows=8
            )
        else:
            df = pd.read_csv(
                f"../Data/{file}",
                encoding="utf-16",
                skiprows=2
            )

        print("\nShape:")
        print(df.shape)

        print("\nColumns:")
        print(list(df.columns))

        print("\nHead(5):")
        display(df.head())

    except Exception as e:
        print("Error:", e)
```

```python
traffic = pd.read_csv("../Data/wc_traffic.csv", skiprows=8)

traffic["Date"] = pd.to_datetime("2026-01-01") + pd.to_timedelta(traffic["Nth day"] - 1, unit="D")

traffic = traffic.rename(columns={
    "Active users": "Active_Users"
})

traffic = traffic[["Date", "Active_Users"]]

display(traffic.head())
print(traffic.shape)
```

```python
ig_views = pd.read_csv("../Data/ig_views.csv", encoding="utf-16", skiprows=2)

ig_views["Date"] = pd.to_datetime(ig_views["Date"])
ig_views = ig_views.rename(columns={"Primary": "IG_Views"})

display(ig_views.head())
print(ig_views.shape)
```

```python
master = pd.merge(
    traffic,
    ig_views,
    on="Date",
    how="left"
)

display(master.head())
print(master.shape)
```

```python
print("IG Views min date:", ig_views["Date"].min())
print("IG Views max date:", ig_views["Date"].max())

print("Traffic min date:", traffic["Date"].min())
print("Traffic max date:", traffic["Date"].max())
```

```python
matching_dates = traffic["Date"].isin(ig_views["Date"]).sum()

print("Traffic rows:", len(traffic))
print("Matching dates with IG views:", matching_dates)
```

```python
def load_meta_file(filename, new_column_name):
    df = pd.read_csv(
        f"../Data/{filename}",
        encoding="utf-16",
        skiprows=2
    )
    df["Date"] = pd.to_datetime(df["Date"])
    df = df.rename(columns={"Primary": new_column_name})
    return df[["Date", new_column_name]]
```

```python
ig_interactions = load_meta_file("ig_interactions.csv", "IG_Interactions")
ig_clicks = load_meta_file("ig_link clicks.csv", "IG_Link_Clicks")
ig_visits = load_meta_file("ig_profile_visits.csv", "IG_Profile_Visits")
ig_follows = load_meta_file("ig_follows.csv", "IG_Follows")

fb_views = load_meta_file("fb_views.csv", "FB_Views")
fb_interactions = load_meta_file("fb_interactions.csv", "FB_Interactions")
fb_clicks = load_meta_file("fb_link clicks.csv", "FB_Link_Clicks")
fb_visits = load_meta_file("fb_visits.csv", "FB_Visits")
fb_follows = load_meta_file("fb_follows.csv", "FB_Follows")
```

```python
master = traffic.copy()

for df in [
    ig_views,
    ig_interactions,
    ig_clicks,
    ig_visits,
    ig_follows,
    fb_views,
    fb_interactions,
    fb_clicks,
    fb_visits,
    fb_follows
]:
    master = pd.merge(master, df, on="Date", how="left")

master = master.fillna(0)

display(master.head())
print(master.shape)
print(master.isna().sum())
```

```python
master["DayOfWeek"] = master["Date"].dt.dayofweek
master["Month"] = master["Date"].dt.month
master["Quarter"] = master["Date"].dt.quarter
master["Weekend"] = (master["DayOfWeek"] >= 5).astype(int)

display(master.head())
print(master.shape)
```

```python
X = master[[
    "IG_Views",
    "IG_Interactions",
    "IG_Link_Clicks",
    "IG_Profile_Visits",
    "IG_Follows",
    "FB_Views",
    "FB_Interactions",
    "FB_Link_Clicks",
    "FB_Visits",
    "FB_Follows",
    "DayOfWeek",
    "Month",
    "Quarter",
    "Weekend"
]]

y = master["Active_Users"]

print(X.shape)
print(y.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

mlr = LinearRegression()

mlr.fit(X, y)

mlr_pred = mlr.predict(X)

mlr_r2 = r2_score(y, mlr_pred)

print("MLR R²:", round(mlr_r2,4))
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

rf.fit(X, y)

rf_pred = rf.predict(X)

rf_r2 = r2_score(y, rf_pred)

print("RF R²:", round(rf_r2,4))
```

```python
top_delta = master.sort_values(
    "Abs_Delta",
    ascending=False
)

display(
    top_delta[
        ["Date",
         "Active_Users",
         "Delta",
         "Abs_Delta"]
    ].head(10)
)
```

```python
master["Predicted_MLR"] = mlr_pred
master["Predicted_RF"] = rf_pred

master["Delta"] = (
    master["Predicted_RF"]
    - master["Predicted_MLR"]
)

master["Abs_Delta"] = abs(master["Delta"])
```

```python
print(master.columns)
```

```python
top_delta = master.sort_values(
    by="Abs_Delta",
    ascending=False
)

display(
    top_delta[
        [
            "Date",
            "Active_Users",
            "Predicted_MLR",
            "Predicted_RF",
            "Delta",
            "Abs_Delta"
        ]
    ].head(10)
)
```
